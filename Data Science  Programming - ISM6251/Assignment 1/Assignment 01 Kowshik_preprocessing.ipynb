About the Dataset:
    
The salaries of employees at a business are detailed in this dataset. The columns include details like age, gender, education level, job title, years of experience, and pay. Each row represents a distinct employee.
Columns:
Age: Each employee's age is shown in years in this section. This column contains numerical numbers.


Gender: This column includes the gender of each employee, which can be either male or female. This column's numbers are categorical.

Education Level: This column lists each employee's educational background, which can range from high school to a PhD or bachelor's degree. This column's numbers are categorical.

employment Title: Each employee's employment title is listed in this column. The job titles can vary depending on the business and may include positions such as manager, analyst, engineer, or administrator. This column's numbers are categorical.

Years of Experience: Each employee's years of job experience are listed in this column. This column contains numerical numbers.

Salary: This column reflects the annual salary of each employee in US dollars. The numeric values in this column can change based on the job title, years of experience, and educational attainment.


The salary is in the form of continuos variable, im converting continous value to discrete value. I'm dropping the records after checking the correlation.
